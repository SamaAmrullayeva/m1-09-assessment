# Strategy: Linear interpolation by city. 
# Justification: PM2.5 levels typically follow temporal trends; 
# interpolation estimates gaps based on surrounding days better than a global mean.

## Part B: Required Analysis (TODOs)

- Data quality analysis: city with highest % invalid/missing pm25.
- Rolling analysis: 7-day rolling average pm25 per city and short explanation.
- Event detection: percentile threshold for high pollution and counts per city.
- Volatility comparison: choose two cities, define a metric, justify result.
- Reshaping: pivot table with months as rows and cities as columns (avg pm25).

```python
# TODO: analysis tasks
# - data quality by city
# 1. City with highest percentage of invalid/missing pm25 (Pre-cleaning check)
raw_pm25 = pd.to_numeric(pd.read_csv('m1-09-assessment.csv')['pm25'], errors='coerce')
city_col = pd.read_csv('m1-09-assessment.csv')['city']
missing_report = pd.DataFrame({'city': city_col, 'pm25': raw_pm25})
highest_missing_city = missing_report.groupby('city')['pm25'].apply(lambda x: x.isnull().mean() * 100).idxmax()

# - rolling averages per city
# 2. 7-day rolling average per city
# Why? It smooths out short-term spikes (like a single high-traffic day) 
# to highlight the broader atmospheric trend.
df = df.sort_index()
df['pm25_7d_roll'] = df.groupby('city')['pm25'].transform(lambda x: x.rolling(window=7).mean())

# - percentile-based event counts
# 3. Percentile-based high-pollution events (e.g., above 90th percentile)
threshold = df['pm25'].quantile(0.90)
high_pollution_counts = df[df['pm25'] > threshold].groupby('city').size()

# - volatility comparison
# 4. Volatility metric: Coefficient of Variation (CV = Std Dev / Mean)
# Justification: CV normalizes volatility, allowing us to compare consistency 
# across cities with different baseline pollution levels.
volatility = df.groupby('city')['pm25'].apply(lambda x: x.std() / x.mean())

# - pivot table by month and city
# 5. Month-by-city pivot table
pivot_pm25 = df.pivot_table(values='pm25', index=df.index.month, columns='city', aggfunc='mean')
```

## Part E: Interpretation Questions (TODOs)

Write short answers here:

1. Which city shows the most persistent high pm25 levels, and what evidence supports that?
The city with the highest median PM2.5 and the most "high-pollution events" (those above your 90th percentile threshold) is the most persistent. This usually happens in cities with dense traffic, heavy industry, or geographic features like valleys that trap stagnant air, preventing pollutants from dispersing.
2. How does missing or invalid data affect your confidence in the results?
Missing data introduces uncertainty bias. If a city is missing values specifically during the winter (when people burn more fuel for heat), your calculated annual average will be artificially lower than the reality. High amounts of missing data mean we can only describe the available data, not the actual environment of the city.
3. Does temperature appear related to pm25 in your analysis? Explain briefly.
Generally, yes. In many European cities, there is an inverse relationship: as temperature drops, PM2.5 often rises. This is due to increased heating emissions and "thermal inversions," where a layer of warm air acts like a lid, trapping cold, polluted air near the ground.
4. What is one limitation of using daily averages for air-quality policy decisions?
Daily averages hide peak exposure levels. For example, air quality might be "Good" on average for 24 hours, but "Hazardous" during the two hours children are walking to school in morning rush-hour traffic. Policy needs to address these dangerous spikes, which daily averages smooth over.
5. If you had one more dataset to improve this analysis, what would it be and why?
A Wind Speed and Direction dataset. Wind is the primary factor in "flushing" pollutants out of a city. Knowing the wind direction would help planners identify if the pollution is being generated locally (e.g., from a specific highway) or being blown in from a neighboring industrial region.
